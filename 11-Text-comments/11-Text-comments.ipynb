Для детекции языка используем Compact Language Detector 2 - он содержит только 80 языков, зато имеет очень высокую производительность и высокую точность детекции - идеальный вариант для детекции наиболее популярных языков.

```python
# подготовленная функция для детекции языка с обработкой исключений
def detect_language(text):
    try:
        _, _, details = cld2.detect(text)
        language_name = details[0][0] if details else 'unknown'
        language_code = details[0][1] if details else 'unknown'
        return language_name, language_code
    except Exception as e:
        return 'unknown', 'unknown'
```

```python
# применяем функцию к столбцу и распаковываем результаты
detection_results = data['text'].apply(
    lambda t: detect_language(t)
)
```

```python
# заполняем список languages и колонку language_code
languages = detection_results.apply(lambda x: x[0]).tolist()
data['language_code'] = detection_results.apply(lambda x: x[1])
```

**Шаг 3:** Запишем функцию для подготовки текста

```python
# инициируем стеммер для стемминга английских слов
stemmer = SnowballStemmer("english")

# подготовим стоп-слова
stop_words = set(stopwords.words('english'))
```

<div class="alert alert-block alert-warning">
Лемматизацию тоже пробовал, но с ней процесс дольше и объем данных больше, при векторизации с большим количеством фич ядро падает, так что упростил предобработку как смог

```python
# функция для подготовки текста
def prepare_text(text):

    # понижение регистра текста
    text = str(text).lower()

    # удаление URL
    text = re.sub(r'http\S+', '', text)

    # удаление HTML тегов
    text = re.sub(r'<.*?>', '', text)

    # удаление специальных символов, оставляем только буквы и пробелы
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # удаление лишних пробелов, если они есть
    text = re.sub(r'\s+', ' ', text).strip()

    # стемминг и удаление стоп-слов
    words = text.split()
    words = [stemmer.stem(word) for word in words if word not in stop_words and len(word) > 2]

    return ' '.join(words)
```

**Шаг 4:** Проведем подготовку текста

Подготовленный текст запишем в отдельную колонку, для его векторизации перед обучением

```python
data['prepared_text'] = data['text'].apply(prepare_text)

# обнулим индексы после предобработки
data = data.reset_index(drop=True)
```

После подготовки текста очистились мусорные комментарии содержащие наборы символов и ip адреса, поэтому индексы обнулили

**Шаг 5:** Мешок слов

```python
# Создадим счётчик для мешка слов
stop_words = stopwords.words('english')
count_vect = CountVectorizer(max_features=40, stop_words=stop_words) # 40 самых популярных слов
```

Создадим мешок слов для каждого класса - это позволит выявить наиболее характерные слова в этих классах

Запишем в переменную токсичные комментарии

```python
toxic_data = data[data['toxic'] == 1]
```

Создадим список самых популярных словофрм для токсичных комментариев

```python
# корпус слов для токсичных комментариев
corpus_toxic = list(toxic_data['prepared_text'])
```

```python
# мешок слов токсичных комментариев
bow_toxic = count_vect.fit_transform(corpus_toxic)
```

```python
# список 40 самых популярных слов
toxic_list = count_vect.get_feature_names_out()
toxic_list
```

Запишем в переменную нетоксичные комментарии

```python
nontoxic_data = data[data['toxic'] == 0]
```

Создадим список самых популярных словоформ для нетоксичных комментариев

```python
# корпус слов нетоксичных комментариев
corpus_nontoxic = list(nontoxic_data['prepared_text'])
```

```python
# мешок слов нетоксичных коментариев
bow_nontoxic = count_vect.fit_transform(corpus_nontoxic)
```

```python
# список самый популярных слов
nontoxic_list = count_vect.get_feature_names_out()
nontoxic_list
```

Видим, что и в токсичной группе и в нетоксичной, есть совпадения по словам, котрые в целом являются нейтральными, их надо будет исключить из списков, так как они будут мешать корректному определению класса.

```python
# общий список совпадающих слов
common_words = list(set(toxic_list) & set(nontoxic_list))

# списки без совпадающих слов
toxic_list = [word for word in toxic_list if word not in common_words]
nontoxic_list = [word for word in nontoxic_list if word not in common_words]
```

Теперь у нас более корректные наборы популярных слов для токсичных и нетоксичных комментариев

**Шаг 6:** Добавление новых признаков

В качестве дополнительных признаков добавим следующие:
 - частота встречи "токсичных" слов в комментарии и частота встречи "нетоксичных" слов в комментарии

```python
# запишем колнку с частотой токсичных слов
data['toxic_freq'] = data['prepared_text'].apply(
    lambda w: count_words_in_list(w, toxic_list)
)

# запишем колонку с частотой нетоксичных слов
data['nontoxic_freq'] = data['prepared_text'].apply(
    lambda w: count_words_in_list(w, nontoxic_list)
)
```

```python
# векторизация текста с комбинацией TF-IDF и CountVectorizer

# векторизация основных признаков текста TF-IDF
tfidf_vectorizer = TfidfVectorizer(
    max_features=8000,
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.7,
    sublinear_tf=True,
    norm='l2'
)

# векторизация дополнительных признаков текста CountVectorizer
count_vectorizer = CountVectorizer(
    max_features=2000,
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.7,
    binary=True
)

X_tfidf_train = tfidf_vectorizer.fit_transform(X_text_train)
X_count_train = count_vectorizer.fit_transform(X_text_train)
X_tfidf_val = tfidf_vectorizer.transform(X_text_val)
X_count_val = count_vectorizer.transform(X_text_val)
X_tfidf_test = tfidf_vectorizer.transform(X_text_test)
X_count_test = count_vectorizer.transform(X_text_test)
```

Подготовим функцию для настройки моделей

```python
# функция для тонкой настройки модели
def optimize_and_evaluate_model(model, param_grid, X_train, y_train, X_val, y_val, model_name):

    # случайный поиск
    search = RandomizedSearchCV(
        model,
        param_grid,
        n_iter=10,
        cv=3,
        scoring='f1',
        n_jobs=-1,
        random_state=RS,
        verbose=0
    )

    # определение времени обучения
    start_time = time.time()
    search.fit(X_train, y_train)
    train_time = time.time() - start_time

    # определение лучшей модели
    best_model = search.best_estimator_

    # определение времени редсказания
    start_time = time.time()
    y_pred = best_model.predict(X_val)
    predict_time = time.time() - start_time

    # определение F1 на валидации
    f1 = f1_score(y_val, y_pred)

    print(model_name)
    print(f"Лучшие параметры: {search.best_params_}")
    print(f"Время обучения: {train_time:.2f} сек")
    print(f"Время предсказания: {predict_time:.2f} сек")
    print(f"F1-score: {f1:.4f}")

    # сохранение лучшей модели в отдельный ключ словаря
    model_var_name = f"best_{model_name.lower()}"
    best_models[model_var_name] = best_model

    return {
        'model': best_model,
        'name': model_name,
        'f1_score': f1,
        'train_time': train_time,
        'predict_time': predict_time
    }
```

```python
# Функция для оценки Dummmy
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    # Обучение модели
    model.fit(X_train, y_train)

    # Предсказания
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None

    # Метрики
    f1 = f1_score(y_test, y_pred)

    print(f"\n{model_name}:")
    print(f"F1-score на тесте: {f1:.4f}")

    return model, f1, y_pred_proba
```

Всего определено 61 язык комментариев. Комментарии на всех языках, кроме английского, составляют около 1,5%. Принято решение оставить только комментарии на английском языке
     - явных дубликатов в данных нет
     - значения тооксичности в норме, их всего 2: 0 и 1. При этом доля нетоксичных комментариев (0) составляет 89,8%, а доля токсичных (1) 10,2% - имеет место дисбаланс классов.
     - распределение длины комментариев таково, что в основной массе нетоксичные комментарии длинее чем токсичные, однако в обоих классах, распределение от 5 символов до 5000, при этом в самых длинных комментариях часто откровенный спам, а в коротких много технического "мусора"
 - в ходе преобработки данных:
     -  удален дубликат идексов, индекс приведены в порядок
     -  удалены лишние языки, оставлены только комментарии на английском языке
     -  сам текст комментариев очищен и подготовлен к векторизации с помощью стемминга
     -  с помощью мешка слов установлены самые популярные слова в токсичных комментариях и в нетоксичных комментариях, на их основе созданы списки слов, с очисткой от дубликатов
     - добавлены новые признаки в датафрейм: частоста использования токсичных и нетоксичных слов в комментарии, эмоциональная окраска комментария с помощью SentimentIntensityAnalyzer.    


<h3>Обучение моделей</h3>
Перед обучением данные подготовлены, проведена комплексная векторизация текста с использованием TF-IDF для основной массы признаков и CountVectorizer для дополнительных признаков.

Для обучения с подбором гиперпараметров были выбраны 4 модели:
- модель логистической регрессии
- модель случайных лесов
- модель градиентного бустинга
- модель опорных векторов с калибровкой

На валидации, со значением стандартного порога классификации ни одна модель не смогла достичь требуемого результата для оценки F1 - 0,75, однако, наиболее близкой к требуемому результату оценки оказалась модель опорных векторов с калибровкой, ее оценка F1 на валидационной выборке составила 0.7416. При этом модель показывает и самое высокое время обучения и достаточно высокое время предсказания

После подбора оптимального порога классификации, установлен порог в 0.3149, при этом результат оценки F1 для выбранной модели на валидационной выборке составил 0.7650, что соответствует требованиям проекта.

При оценке на тестовой выборке с установлением порога классификации, результат оценки F1 составил 0.7780, время предсказания составило 0.01 секунды

**По результатам работы, для класификации текстов на токсичные и нетоксичные принято решение о выборе модели опорных векторов с калибровкой и установлением порога классификации в 0.3149.**
