## Анализ

### Анализ средних значений

 Для анализа построим графики по разным срезам данных

```python
# сделаем копию датафрейма для построения графика средних значений
data_mean = data.copy()

# построим график
plt.figure(figsize=(10, 5))
# для уменьшения шума ресемплируем по неделям
data_mean.resample('1W').mean().plot(ax=plt.gca(), grid=True)
plt.title('Средние значения')
plt.legend('Количество заказов', loc='upper left')
plt.xlabel('Временной период')
plt.ylabel('Количество заказов')
plt.show()
```

Анализ средних значений, при ресемплировании понедельно, во избежание шумов, показывает обший восходящий тренд по количеству заказов, самый резкий рост отмечен в течение августа

### Анализ скользящего среднего

```python
# скопируем датафрейм после ресемлирования для построения графика по новым признакам
data_plot = data.copy()

# создадим признак скользящего среднего
data_plot['rolling_mean'] = data_plot.shift(1).rolling(24).mean()

# создадим признак скользящего стандартного отклонения
data_plot['rolling_std'] = data_plot[['num_orders']].rolling(24).std()
```

```python
# построим сводный график с новыми признаками
plt.figure(figsize=(10, 5))
data_plot.plot(ax=plt.gca(), grid=True)
plt.title('Cкользящее среднего и стандартного отклонения')
plt.legend(['Количество заказов', 'Скользящее среднее отклонение', 'Скользящее стандарное отклонение'])
plt.xlabel('Временной период')
plt.ylabel('Количество заказов')
plt.show()
```

По общему графику заказов, скользящему среднему отклонению и скользящему стандартному отклонению видно что в целом количество заказов увеличивается, однако именно в в августе как по всем трем показателям видно резуое увеличение, так и, в том числе, заметны резкие всплески в виде пиков.

В целом ряды можно охарактеризовать как нестационарные, так как у них отсутствуют признаки периодичности.

Однако стоит посмотреть на разности временного ряда, с помощью метода shift

```python
# создадим признак разности
data_plot['orders_shift'] = data_plot['num_orders'] - data_plot['num_orders'].shift()
# пересчитаем скользящее среднего отклонения по признаку разности
data_plot['rolling_mean'] = data_plot['orders_shift'].rolling(10).mean()
# пересчитаем скользящее стандартного отклонения по признаку разности
data_plot['rolling_std'] = data_plot['orders_shift'].rolling(10).std()

# построим график по новым признакам
plt.figure(figsize=(10, 5))

data_plot['orders_shift'].plot(ax=plt.gca(), grid=True)
data_plot['rolling_mean'].plot(ax=plt.gca())
data_plot['rolling_std'].plot(ax=plt.gca())

plt.title('Разность времменных рядов')
plt.legend(['Разность количество заказов', 'Скользящее среднее отклонение', 'Скользящее стандарное отклонение'])
plt.xlabel('Временной период')
plt.ylabel('Количество заказов')
plt.show()
```

Видим, что на разности ряды стали более стационарными по всем трем показателям. Да, есть точечные всплески, но в целом все три ряда показывают признаки стационарности

### Анализ тренда, сезонности и остатков

```python
# проведем декомпозицию
decomposed = seasonal_decompose(data.resample('1D').sum())

# построим график по признакам декомпозиции
plt.figure(figsize=(10, 15))

plt.subplot(311)
# график тренда
decomposed.trend.plot(ax=plt.gca(), grid=True)
plt.title('Тренд')
plt.xlabel('Временной период')
plt.ylabel('Количество заказов')

plt.subplot(312)
# график сезонности
decomposed.seasonal.plot(ax=plt.gca(), grid=True)
plt.title('Сезонность')
plt.xlabel('Временной период')
plt.ylabel('Распределение')

plt.subplot(313)
# график остатков
decomposed.resid.plot(ax=plt.gca(), grid=True)
plt.title('Остатки')
plt.xlabel('Временной период')
plt.ylabel('Распределение')

plt.tight_layout()
```

Видим, что общий тренд восходящий, не смотря не набольшие спады, в последний месяц рост заказов по тренду увеличился.

При помесячном анализе сезонности видим, наблюдается сезонность внутри месяца, при подневном разрезе,
Поскольку мы изучаем срез лишь за часть одного года, помесячную сезонность в разрезе сравнения разных годов изучить не получится. Отдельно изучим сезонность внури месяца, в сравнении разных месяцев и подневно в разрезе сравнения разных недель месяца.

В целом, сезонность показывает признаки стационарного ряда.

Анализ остатков показывает неоднородность, с разным распределением пиков в разные месяцы. Однако наибольшие колебания значений мы видим в марте, мае, июне и августе.

При этом график автокореляции показывает, что


### Месячный анализ

Для месячного анализа, необходимо сделать срез по месяцу. Для корректности сделаем два среза:
 - за март, как первый месяц
 - за август, последний месяц, тем более что динамика августа оличается от динамики марта

```python
# проведем декомпозицию для данных марта
decomposed_march = seasonal_decompose(data['2018-03-01':'2018-03-31'].resample('1d').sum())
# проведем декомпозицию для данных августа
decomposed_august = seasonal_decompose(data['2018-08-01':'2018-08-31'].resample('1d').sum())

# построим графики для месячных данных
plt.figure(figsize=(12, 12))

# графики тренда, сезонности и остатков для марта
plt.subplot(321)
decomposed_march.trend.plot(ax=plt.gca(), grid=True)
plt.title('Тренд марта')
plt.xlabel('Временной период')
plt.ylabel('Количество заказов')
plt.subplot(323)
decomposed_march.seasonal.plot(ax=plt.gca(), grid=True)
plt.title('Сезонность марта')
plt.xlabel('Временной период')
plt.ylabel('Распределение')
plt.subplot(325)
decomposed_march.resid.plot(ax=plt.gca(), grid=True)
plt.title('Остатки марта')
plt.xlabel('Временной период')
plt.ylabel('Распределение')

# графики тренда, сезонности и остатков для августа
plt.subplot(322)
decomposed_august.trend.plot(ax=plt.gca(), grid=True)
plt.title('Тренд августа')
plt.xlabel('Временной период')
plt.ylabel('Количество заказов')
plt.subplot(324)
decomposed_august.seasonal.plot(ax=plt.gca(), grid=True)
plt.title('Сезонность августа')
plt.xlabel('Временной период')
plt.ylabel('Распределение')
plt.subplot(326)
decomposed_august.resid.plot(ax=plt.gca(), grid=True)
plt.title('Остатки августа')
plt.xlabel('Временной период')
plt.ylabel('Распределение')

plt.tight_layout()
```

При месячном анализе видно, что тренд марта хоть и восходящий, но неоднородный, с сильными падениями и ростом: в разные дни количество заказов разное и бессистемное. При этом тренд августа более равномерный и восходящий.

Подневная сезонность видна и в мартовском и в августовском срезах, однако картина сезонности разная.

Остатки тоже имеют разную и несистемную картину, являясь нестационарными

## Обучение

### Добавление признаков

Запишем функцию для добавления новых признаков

```python
def make_features(data, max_lag, rolling_mean_size):
    mf_data = data.copy()
    mf_data['day'] = mf_data.index.day
    mf_data['dayofweek'] = mf_data.index.dayofweek
    mf_data['hour'] = mf_data.index.hour

    for lag in range(1, max_lag + 1):
        mf_data['lag_{}'.format(lag)] = mf_data['num_orders'].shift(lag)

    mf_data['rolling_mean'] =  mf_data['num_orders'].shift().rolling(rolling_mean_size).mean()
    mf_data.dropna(inplace=True) # обязательно удалить пропуски для обучения!
    return mf_data
```

Теперь запишем новые признаки с помощью функции. Для максимального запаздывания - max_lag и окна скользящего среднего - rolling_mean_size зададим значение 24 - так как в сутках 24 часа, а у нас ресемплирование и прогнозирование по часам.

```python
new_data = make_features(data, 24, 24)
new_data
```

### Разбивка данных

Проведем разбивку на данных на входные признаки и целевой

```python
# выделим входные признаки
X = new_data.drop(['num_orders'], axis=1)
# выделим целевой признак
y = new_data['num_orders']
```

```python
# сделаем разбивку на тренировочную и валидационную выборки
# пропорция 8:2
X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    shuffle=False,
    test_size=0.2)

# выделим из валидационной тестовую выборку
# пропорция 1:1
X_valid, X_test, y_valid, y_test = train_test_split(
    X,
    y,
    shuffle=False,
    test_size=0.5)
# по итогу у нас пропорции 8:1:1
```

### Подготовка

Подготовим датафрейм для записи результатов обучения моделей

```python
results = pd.DataFrame(columns=['model', 'train_ time', 'pred_time', 'RMSE'])
```

Подготовим метод кросс валидации для временных рядов

```python
tscv = TimeSeriesSplit(
    n_splits=3, # 3 сплита так как в трейне 3,5 к строк
    test_size=1000, # размер по 1000
    gap=0
)
```

Подготовим функцию для подбора гиперпараметров и фиксации результатов обучения

```python
def fit_model(estimator, param_grid, X_train, y_train, X_valid, y_valid):
    model = GridSearchCV(estimator=estimator,
                         param_grid=param_grid,
                         n_jobs=-1,
                         cv=tscv,
                         scoring='neg_root_mean_squared_error'
                         )
# время обучения
    start_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - start_time

# лучшая оценка
    best_score = abs(round(model.best_score_, 1))

    print(f'Лучшее значение оценки: {best_score}')
    print(f'Лучшие параметры: {model.best_params_}')

# лучшая модель
    best_model = estimator.set_params(**model.best_params_)

# время предсказания
    start_time = time.time()
    best_model.fit(X_train, y_train)
    prediction_time = time.time() - start_time

# RMSE на валидации
    rmse = mean_squared_error(y_valid, best_model.predict(X_valid), squared=False).round(8)

    return best_model, best_score, training_time, prediction_time, rmse
```

### Обучение модели линейной регрессии

```python
# передадим функции параметры для линейной регрессии
best_model_LR, best_score, training_time, prediction_time, rmse_LR = fit_model(
    LinearRegression(),
    [{'fit_intercept': [True, False]},
     {'copy_X': [True, False]},
     {'n_jobs': [1, -1]}],
    X_train,
    y_train,
    X_valid,
    y_valid
)
```

```python
# запишем результаты
results.loc[0] = ['LinearRegression', training_time, prediction_time, rmse_LR]
results
```

```python
# запишем результаты
results.loc[1] = ['DecisionTreeRegressor', training_time, prediction_time, rmse_DTR]
results
```
